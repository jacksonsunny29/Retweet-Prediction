```python
# Importing the relevant modules
from transformers import BertTokenizer, BertModel, BertForSequenceClassification
import pandas as pd
import numpy as np
import torch
import csv
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
```

```python
# Load the tokenizer of the "bert-base-cased" pretrained model
# See https://huggingface.co/transformers/pretrained_models.html for other models
tz = BertTokenizer.from_pretrained("bert-base-uncased")
```

```python
x=pd.read_csv("/home/bruce/EP-M1/ML-DL/Project/train.csv")
```

```python
x_text=x['text']
x_vect=[]
```

```python
for i in x_text:
    sent = i
    encoded = tz.encode_plus(
    text=sent,  # the sentence to be encoded
    add_special_tokens=True,  # Add [CLS] and [SEP]
    max_length = 35,  # maximum length of a sentence
    pad_to_max_length=True,  # Add [PAD]s
    return_attention_mask = True,  # Generate the attention mask
    return_tensors = 'pt',  # ask the function to return PyTorch tensors
    )
    input_ids = encoded['input_ids'] 
    x_vect.append(input_ids)
```

```python
x_vec=[]
for i in x_vect:
    i = torch.flatten(i)
    i = i.numpy()
    x_vec.append(i)
```

```python
x_vec
```

```python
x["tv"]=x_vec
```

```python
x = x.drop(["text","mentions","urls","hashtags","timestamp","TweetID"],axis=1)
```

```python
y_vec =[]
for i in y_vect:
    i = torch.flatten(i)
    i = i.numpy()
    y_vec.append(i)
```

```python
X = x[["favorites_count","followers_count","statuses_count","friends_count","verified"]]
X = pd.concat([X,x["tv"]])
```

```python
X
```

```python
Y = x["retweets_count"]
```

```python
# Set the model's hyperparameters
learning_rate = 3e-5
batch_size = 8
num_epochs = 100

# Create the neural network model
model = Sequential()
model.add(Dense(1))
model.add(Dense(1))
model.add(Dense(1))

# Compile and train the model
model.compile(
    optimizer='adam',
    loss='MeanAbsoluteError'
)
model.fit(
    X, Y,
    batch_size=batch_size,
    epochs=num_epochs
)

# Use the trained model to make predictions
predictions = model.predict(X)
```
